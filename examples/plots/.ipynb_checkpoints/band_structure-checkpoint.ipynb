```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_palette(sns.color_palette("RdBu", n_colors=8, desat=1.))
sns.set_style("white")
sns.set_style({'grid.color': '0.95'})
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']
```

```python
bands = np.loadtxt("../data/bands.csv", delimiter = ',')
```

```python
NK = 512
NA = 3
fig = plt.figure(1, figsize = (6, 4))
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.linspace(-np.pi / NA, np.pi / NA, num=NK + 1,\
        endpoint=False), bands, linewidth = 0.8, color = 'darkblue')
ax.set_xlabel(r'$k$', fontsize = 12)
ax.set_ylabel(r'$\varepsilon_{k, \sigma} [eV]$', rotation = 0,\
fontsize = 12, labelpad = 30)
ax.set_xticks([-np.pi / NA, -np.pi / NA / 2, 0, np.pi / NA / 2, np.pi / NA])
if NA == 1:
    ax.set_xticklabels([r'$-\pi$', r'$0$', r'$\pi$'])
else:
    ax.set_xticklabels([r'$-\pi / ' + str(NA) + '$', r'$-\pi / ' + str(NA * 2) + '$',\
                        r'$0$', r'$\pi / ' + str(NA * 2) + '$', r'$\pi / ' + str(NA) + '$'])
```

```python
nUp = np.loadtxt("../data/nUp.csv", delimiter = ',')
```

```python
nDw = np.loadtxt("../data/nDw.csv", delimiter = ',')
```

```python
NORB = 3
NY = 6
NA = 4
```

```python
fig = plt.figure(1, figsize=(12, 8))
nx=12
dotscale = 500
clr = np.chararray((nx * NY), itemsize = 10)
a1 = np.arange(nx)
a2 = np.arange(NY)
lat = np.zeros((nx * NY))
vs = np.zeros((nx * NY , 2))
v1 = np.array([1, 0])
v2 = np.array([1 / 2, np.sqrt(3) / 2])
for i in range(nx):
    for j in range(NY):
        vs[nx * j + i, :] =\
        a1[i] * v1 + a2[j] * v2
        a = (i + NA) % NA
        lat[nx * j + i] \
        = ( nUp[NORB * (NY * a + j)]\
        + nUp[1 + NORB * (NY * a + j)]\
        + nUp[2 + NORB * (NY * a + j)]\
        - nDw[NORB * (NY * a + j)]\
        - nDw[1 + NORB * (NY * a + j)]\
        - nDw[2 + NORB * (NY * a + j)])
        if lat[nx * j + i] > 0:
            clr[nx * j + i] = "#95a5a6"
        else:
            clr[nx * j + i] = "#e74c3c"
print(lat[0], lat[1])
print(lat[nx* (NY - 1)], lat[nx* (NY - 1) + 1])

ax = fig.add_subplot(111)
ax.axis('off')
ax.scatter(vs[:, 0], vs[:, 1],\
s = dotscale * abs(lat),\
c = clr.decode('UTF-8'), alpha = 0.8, edgecolors = None)
ax.axis('equal')
ax.set_yticklabels([])
ax.set_xticklabels([])
```
