```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_palette(sns.color_palette("RdBu", n_colors=8, desat=1.))
sns.set_style("white")
sns.set_style({'grid.color': '0.95'})
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']
```

```python
TBbandsUp = np.loadtxt("free-bands-up.csv")
TBbandsDw = np.loadtxt("free-bands-dw.csv")
```

```python
NK = 512

fig = plt.figure(2, figsize = (9, 6))
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.linspace(-np.pi, np.pi, num=NK,\
        endpoint=False), TBbandsUp, linewidth = 0.8, color = 'darkblue')
ax.plot(np.linspace(-np.pi, np.pi, num=NK,\
        endpoint=False), TBbandsDw, linewidth = 0.8, color = 'darkred')
ax.set_xlabel(r'$k$', fontsize = 12)
ax.set_ylabel(r'$\varepsilon_{k, \sigma} [eV]$', rotation = 0,\
fontsize = 12, labelpad = 30)
```
